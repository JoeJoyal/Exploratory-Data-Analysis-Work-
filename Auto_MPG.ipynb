```python
from sklearn.cross_validation import train_test_split
from sklearn import ensemble
from sklearn.metrics import mean_absolute_error
from sklearn.externals import joblib

df=pd.read_csv('auto-mpg.csv')

del df['car name']
del df['acceleration']
del df['horsepower']
del df['model year']

feature_df=pd.get_dummies(df,columns=['origin','cylinders'])

del feature_df['mpg']

x=feature_df.as_matrix()
y=df['mpg'].as_matrix()

x,y
```

```python
feature_df.to_csv('SampleA.csv')
df['mpg'].to_csv('SampleB.csv')
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3)
x_train, x_test, y_train, y_test
```

```python
model =ensemble.GradientBoostingRegressor(
    n_estimators=1000,
    learning_rate=0.1,
    max_depth=6,
    min_samples_leaf=9,
    max_features=0.1,
    loss='huber',
    random_state=0)
model.fit(x_train,y_train)
```

```python
joblib.dump(model,'model/trained_auto-mpg_classifier_model_GBR.pkl')
```

```python
mae = mean_absolute_error(y_train,model.predict(x_train))
print('Training set mean_absolute_error: %.4f' %mae)

mae = mean_absolute_error(y_test,model.predict(x_test))
print('Test set mean_absolute_error: %.4f' %mae)
```

```python
import pandas as pd
from sklearn.cross_validation import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn import linear_model
from sklearn.linear_model import SGDRegressor
from sklearn import ensemble
from sklearn.metrics import mean_absolute_error
from sklearn.externals import joblib

# Load the data set
df =pd.read_csv('auto-mpg.csv')

#Remove the field from the data set that we don't want to include in our model
del df['car name']
del df['displacement']
del df['acceleration']
del df['model year']
del df['horsepower']
del df['weight']

#Replace the categorical data with one hot coding
features_df =pd.get_dummies(df,columns=['cylinders','origin'])

#Remove the miles per gallon from feature data
del features_df['mpg']

#Create the x and Y
x= features_df.as_matrix()
y=df['mpg'].as_matrix()

#split the data set from training set (70%) to test set (30%)
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3)

#fit regression model
model = ensemble.GradientBoostingRegressor()
model.fit(x_train,y_train)

#save the train model to a file so we can use it another program
joblib.dump(model,'model/training_auto-mpg_model3.pkl')

#Find the error rate on train model
mae = mean_absolute_error(y_train, model.predict(x_train))
print('Training set Mean absolute error :%.4f' %mae)

#Find the error rate on test model
mae = mean_absolute_error(y_test, model.predict(x_test))
print('Test set Mean absolute error: %.4f' %mae)
```

```python
train_val=pd.DataFrame({"Actual":y_train,"Pred":model.predict(x_train)})
test_val=pd.DataFrame({"Actual":y_test,"Pred":model.predict(x_test)})
test_val.head()
```
